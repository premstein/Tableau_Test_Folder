```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,10,10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(10,18))
plt.subplot(4,2,1)
plt.plot(y1,'gX')
plt.xlabel('$y_1 = x$')

plt.subplot(4,2,2)
plt.plot(y2,'bX')
plt.xlabel('$y_2 = x^2$')

plt.subplot(4,2,3)
plt.plot(y3,'rX')
plt.xlabel('$y_3 = x^3$')

plt.subplot(4,2,4)
plt.plot(y4,'yX')
plt.xlabel('$y_4 = \sqrt {x}$')
```

```python
import pandas as pd 
x = pd.read_csv('F:/GITHUB/Python/Salary.csv')
x.head(3)
```

```python
age = x.Age
sal_all = x.All_Devs
sal_pyth = x.Python
sal_java = x.JavaScript
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(7,5),dpi=100)
plt.plot(age,sal_all,'r--',label='sal_all')
plt.plot(age,sal_pyth,'g--',label='sal_python')
plt.plot(age,sal_java,'b--',label='sal_java')
plt.xlabel('Age',fontsize=8)
plt.ylabel('Salary',fontsize=8)
plt.tight_layout()
plt.legend(fontsize=10)
```
